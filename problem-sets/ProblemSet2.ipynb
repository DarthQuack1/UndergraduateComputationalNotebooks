# Vector Operations

## Problem 1: Economic Quantities as Vectors

Suppose we have data on five countries. Their GDP values (in trillions
of USD) are:

$$
\text{gdp} = \begin{pmatrix} 25.5 \\ 18.0 \\ 4.2 \\ 4.1 \\ 3.1 \end{pmatrix}
$$

and their consumption shares of GDP are:

$$
\text{cshare} = \begin{pmatrix} 0.68 \\ 0.54 \\ 0.55 \\ 0.52 \\ 0.65 \end{pmatrix}
$$

**(a)** Create these two vectors in Julia. Compute the consumption
vector $c = \text{gdp} \odot \text{cshare}$ (element-wise
multiplication). What is the total consumption across all five
countries?

```julia
gdp    = [25.5, 18.0, 4.2, 4.1, 3.1]
cshare = [0.68, 0.54, 0.55, 0.52, 0.65]

c = gdp .* cshare
total_consumption = sum(c)

println("Consumption vector c = ", c)
println("Total consumption = ", total_consumption, " trillion USD")
```

**(b)** Compute the dot product $\text{gdp} \cdot \text{cshare}$. Verify
that it equals the sum of the consumption vector from part (a). Explain
economically why $\text{gdp} \cdot \text{cshare} = \sum_i c_i$.

```julia
dot_product = gdp ⋅ cshare   

println("gdp ⋅ cshare  = ", dot_product)
println("sum(c) = ", sum(c))
println("Equal? ", dot_product ≈ sum(c))

println("The dot product gdp · cshare = Σᵢ gdpᵢ × cshareᵢ, which is exactly the sum of each country's consumption (gdpᵢ × cshareᵢ = cᵢ). 
So it's just adding up every country's consumption — which is the same as summing the consumption vector c.")
```

**(c)** Compute the Euclidean norm $\|\text{gdp}\|$ and the norm
$\|c\|$. Which is larger, and why must this always be the case when all
consumption shares are less than 1?

```julia
using LinearAlgebra

norm_gdp = norm(gdp)
norm_c   = norm(c)

println("‖gdp‖ = ", norm_gdp)
println("‖c‖   = ", norm_c)
println("‖gdp‖ > ‖c‖ ? ", norm_gdp > norm_c)

# Since every cshareᵢ ∈ (0,1), each cᵢ = gdpᵢ × cshareᵢ < gdpᵢ,
# so every element shrinks, making ‖c‖ < ‖gdp‖ always.
```

## Problem 2: Projections

The **projection** of vector $a$ onto vector $b$ is defined as:

$$
\text{proj}_b(a) = \frac{a \cdot b}{\|b\|^2} \, b
$$

This gives the component of $a$ that lies in the direction of $b$.

**(a)** Let $a = [3.0, 4.0]$ and $b = [1.0, 0.0]$. Compute
$\text{proj}_b(a)$ by hand and verify using Julia.

```julia
a = [3.0, 4.0]
b = [1.0, 0.0]

proj = (a ⋅ b / (b ⋅ b)) .* b

println("proj_b(a) = ", proj)
# By hand: (3·1 + 4·0)/(1²+0²) × [1,0] = 3 × [1,0] = [3, 0]
```

**(b)** Now let $a = [6.0, 2.0, 3.0]$ and $b = [1.0, 1.0, 1.0]$. Compute
the projection of $a$ onto $b$. Verify that the residual
$r = a - \text{proj}_b(a)$ is orthogonal to $b$ (i.e., $r \cdot b = 0$).

> **Hint**
>
> This orthogonality condition is the same principle behind OLS
> regression: the residuals are orthogonal to the regressors.

```julia
a = [6.0, 2.0, 3.0]
b = [1.0, 1.0, 1.0]

proj = (a ⋅ b / (b ⋅ b)) .* b
r    = a .- proj

println("proj_b(a) = ", proj)
println("residual r = ", r)
println("r ⋅ b = ", r ⋅ b, "  (should be ≈ 0)")
```

# Matrix Operations

## Problem 3: Input-Output Tables

Consider a simple economy with three sectors: Agriculture (A),
Manufacturing (M), and Services (S). The **input-output matrix** records
how much each sector purchases from every other sector per dollar of
output:

$$
A = \begin{pmatrix}
0.10 & 0.25 & 0.05 \\
0.20 & 0.15 & 0.10 \\
0.05 & 0.10 & 0.20
\end{pmatrix}
$$

Here $A_{ij}$ is the amount of sector $i$’s output used as input by
sector $j$ per dollar of $j$’s output.

**(a)** Create the matrix $A$ in Julia. Compute $A'$ (the transpose). Is
$A$ symmetric?

```julia
A = [0.10 0.25 0.05;
     0.20 0.15 0.10;
     0.05 0.10 0.20]

A_T = A'

println("A =\n", A)
println("A' =\n", A_T)
println("A symmetric? ", A == A_T)
```

**(b)** Compute the column sums of $A$. What do these represent
economically? Which sector has the highest intermediate input
requirements?

> **Hint**
>
> Use `sum(A, dims=1)` to sum along rows (giving column sums).

```julia
col_sums = sum(A, dims=1)
println("Column sums: ", col_sums)
# Each column sum = total intermediate input cost per dollar of that sector's output.
# The sector with the largest column sum has the highest intermediate input requirements.
println("Highest intermediate input sector: ", argmax(vec(col_sums)))
```

**(c)** Suppose the final output vector is $x = [100, 200, 300]'$ (in
millions of dollars). Compute the total intermediate demand $Ax$.

```julia
x = [100.0, 200.0, 300.0]
intermediate_demand = A * x
println("Intermediate demand Ax = ", intermediate_demand)
```

## Problem 4: Matrix Multiplication

**(a)** Create the matrices:

$$
B = \begin{pmatrix} 1 & 2 \\ 3 & 4 \end{pmatrix}, \qquad
C = \begin{pmatrix} 5 & 6 \\ 7 & 8 \end{pmatrix}
$$

Compute $BC$ and $CB$. Are they equal? Why or why not?

```julia
B = [1 2; 3 4]
C = [5 6; 7 8]

BC = B * C
CB = C * B

println("BC =\n", BC)
println("CB =\n", CB)
println("BC == CB ? ", BC == CB)
# Matrix multiplication is generally not commutative.
```

**(b)** Compute $(BC)'$ and $C'B'$. Verify that they are equal. This is
a fundamental property of matrix transposition: $(BC)' = C'B'$.

```julia
println("(BC)' =\n", BC')
println("C'B'  =\n", C' * B')
println("Equal? ", BC' == C' * B')
```

**(c)** Compute $B \circ C$ (element-wise multiplication using `.*`).
Compare this to $BC$. Are they different?

```julia
hadamard = B .* C

println("B ∘ C (elementwise) =\n", hadamard)
println("BC (matrix product) =\n", BC)
println("Same? ", hadamard == BC)
```

# Solving Linear Systems

## Problem 5: Market Equilibrium

Consider a two-good economy. The market clearing conditions require:

$$
\begin{aligned}
2p_1 + p_2 &= 10 \\
p_1 + 3p_2 &= 15
\end{aligned}
$$

where $p_1$ and $p_2$ are the equilibrium prices.

**(a)** Write this system in matrix form $Ax = b$. Create the matrix $A$
and vector $b$ in Julia.

```julia
M = [2.0 1.0; 1.0 3.0]
b5 = [10.0, 15.0]

println("M =\n", M)
println("b = ", b5)
```

**(b)** Verify that $A$ is invertible by computing $\det(A)$.

```julia
d = det(M)
println("det(M) = ", d)
println("Invertible? ", d ≠ 0)
```

**(c)** Solve for the equilibrium prices using the backslash operator
`A \ b`. Verify your answer by computing $A x$ and checking that it
equals $b$.

```julia
p = M \ b5
println("Equilibrium prices p = ", p)
println("Verification Mp = ", M * p, "  (should equal b)")
```

**(d)** Also solve using the inverse: $x = A^{-1} b$. Verify that you
get the same answer.

```julia
p_inv = inv(M) * b5
println("p via M⁻¹b = ", p_inv)
println("Same as backslash? ", p ≈ p_inv)
```

## Problem 6: Leontief Model

The **Leontief input-output model** determines the total output $x$
needed to satisfy a vector of final demands $d$, accounting for the fact
that producing output requires intermediate inputs. The relationship is:

$$
x = Ax + d
$$

where $A$ is the input-output matrix from Problem 3 and $d$ is the final
demand vector. Rearranging:

$$
(I - A) x = d
$$

**(a)** Suppose final demand is $d = [50, 80, 120]'$ (millions of
dollars). Construct the matrix $(I - A)$ where $I$ is the $3 \times 3$
identity matrix. Verify that $(I - A)$ is invertible.

> **Hint**
>
> Julia provides a special identity matrix object `I` that automatically
> conforms to the correct size, so you can simply write `I - A`.

```julia
IminusA = I - A
println("(I - A) =\n", IminusA)
println("det(I - A) = ", det(IminusA))
println("Invertible? ", det(IminusA) ≠ 0)
```

**(b)** Solve for the total output vector $x = (I - A)^{-1} d$ using the
backslash operator. How much must each sector produce in total
(including intermediate goods)?

```julia
d6 = [50.0, 80.0, 120.0]
x6 = IminusA \ d6
println("Total output x = ", x6, " (millions USD)")
println("  Agriculture: ", x6[1])
println("  Manufacturing: ", x6[2])
println("  Services: ", x6[3])
```

**(c)** Compute the **Leontief inverse** $L = (I - A)^{-1}$ explicitly.
The element $L_{ij}$ tells you the total output of sector $i$ required
(directly and indirectly) to deliver one dollar of final demand from
sector $j$. Which sector has the largest total output multiplier
(largest column sum of $L$)?

```julia
L = inv(IminusA)
multipliers = sum(L, dims=1)
println("Leontief inverse L =\n", L)
println("Output multipliers (column sums): ", multipliers)
println("Largest multiplier: sector ", argmax(vec(multipliers)))
```

# Economic Application: Input-Output Model

## Problem 10: The Leontief Model in Depth

Consider a four-sector economy (Agriculture, Manufacturing, Energy,
Services) with the following technology matrix:

$$
A = \begin{pmatrix}
0.05 & 0.15 & 0.10 & 0.02 \\
0.10 & 0.20 & 0.15 & 0.08 \\
0.08 & 0.10 & 0.05 & 0.05 \\
0.03 & 0.05 & 0.10 & 0.10
\end{pmatrix}
$$

The element $A_{ij}$ represents how many dollars of sector $i$ output
are needed to produce one dollar of sector $j$ output.

**(a)** Create the matrix $A$ in Julia. Compute the column sums. What
fraction of each sector’s gross output goes to intermediate use (as
opposed to final demand)?

```julia
A10 = [0.05 0.15 0.10 0.02;
       0.10 0.20 0.15 0.08;
       0.08 0.10 0.05 0.05;
       0.03 0.05 0.10 0.10]

col_sums10 = sum(A10, dims=1)
println("Column sums (intermediate input shares): ", col_sums10)
# Each value is the fraction of that sector's gross output used as intermediate inputs.
```

**(b)** Verify that the Leontief model is well-defined by checking that
all eigenvalues of $A$ are less than 1 in absolute value. This is a
necessary condition for the Leontief inverse $(I - A)^{-1}$ to exist and
have non-negative entries.

> **Hawkins-Simon Condition**
>
> If all eigenvalues of $A$ are strictly inside the unit circle
> ($|\lambda_i| < 1$), the economy is productive: it can produce
> positive net output in every sector. If any eigenvalue equals or
> exceeds 1, the economy requires more inputs than it produces.

```julia
eigenvalues = eigvals(A10)
println("Eigenvalues of A: ", eigenvalues)
println("All |λ| < 1? ", all(abs.(eigenvalues) .< 1))
```

**(c)** Compute the Leontief inverse $L = (I - A)^{-1}$. Find the output
multiplier for each sector (the column sums of $L$). Which sector has
the largest multiplier? Interpret this economically.

```julia
L10 = inv(I - A10)
mult10 = sum(L10, dims=1)
println("Output multipliers: ", mult10)
sectors = ["Agriculture", "Manufacturing", "Energy", "Services"]
println("Largest multiplier: ", sectors[argmax(vec(mult10))])
# Lᵢⱼ = total output of sector i needed per dollar of final demand for sector j.
```

**(d)** Suppose final demand is $d = [30, 60, 40, 100]'$ (millions of
dollars). Solve for the total output required from each sector. How much
gross output does Manufacturing need to produce?

```julia
d10 = [30.0, 60.0, 40.0, 100.0]
x10 = (I - A10) \ d10
for (s, v) in zip(sectors, x10)
    println("  $s: $(round(v, digits=2)) million USD")
end
```

**(e)** **Demand shock analysis.** Now suppose final demand for Services
increases by 20% (from 100 to 120 million), while all other demands
remain unchanged. Compute the new total output vector. By how much does
each sector’s output increase? Which sector is most affected (in
absolute terms) by the increase in Services demand?

```julia
d10_shock = [30.0, 60.0, 40.0, 120.0]   # Services demand up 20%
x10_shock = (I - A10) \ d10_shock
Δx = x10_shock .- x10

println("Change in output by sector:")
for (s, v) in zip(sectors, Δx)
    println("  $s: +$(round(v, digits=2)) million USD")
end
println("Most affected: ", sectors[argmax(Δx)])
```

**(f)** **Structural change.** Suppose a technological improvement
reduces Manufacturing’s input requirement from Agriculture by half:
$A_{1,2}$ drops from 0.15 to 0.075. Recompute the Leontief inverse and
the output multiplier for Manufacturing. How does the technological
change affect the Manufacturing multiplier?

```julia
A10_new = copy(A10)
A10_new[1, 2] = 0.075   # Agriculture input to Manufacturing halved

L10_new = inv(I - A10_new)
mult10_new = sum(L10_new, dims=1)

println("Old Manufacturing multiplier: ", round(mult10[2], digits=4))
println("New Manufacturing multiplier: ", round(mult10_new[2], digits=4))
println("Change: ", round(mult10_new[2] - mult10[2], digits=4))
# Reducing input requirements lowers the multiplier — less indirect demand is generated.
```
